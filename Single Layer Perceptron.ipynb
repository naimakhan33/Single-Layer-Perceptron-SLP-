```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions


dataset = pd.read_csv(r"C:\Users\Ahsan\Downloads\large_linearly_separable_placement_dataset (1).csv")
dataset.head(5)


```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="CGPA" , y="Test_Score" , data=dataset , hue="Placement")
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split

x_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100 , pr.score(X_test,y_test)
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=pr)
plt.show()
```
